### Scientific Question: If mice and humans have different Per2 gene protein sequences, will this affect the expression of the Per2 gene in different tissues? 

>The Per2 gene is part of the Period gene family and participates in the regulation of the circadian clock. The Per2 gene is the main component of the mammilian pacemaker. Mutations in this gene are linked to an altered circadian period and the formation of tumors. The Per2 gene is found in many mammals but was first discovered in the circadian clock of Drosophila. https://en.wikipedia.org/wiki/PER2

>The Per2 gene interacts with the Cry gene and forms a heterodimer. After forming a heterodimer it acts to inhibit its own transcription by binding and suppressing the CLOCK/BMAL1 complex which results in a negative feedback loop https://en.wikipedia.org/wiki/PER2. The Per2 gene has been studied in many different forms and the gene sequence, protien sequence and transcription sequence can all be found on the NCBI database under many different organisms that contain the gene. https://www.ncbi.nlm.nih.gov/gene/?term=Per2

>NCBI is a website that includes many different databases that are useful for bioinformatics tools. On these databases you can find different genes, proteins, biomedical literature, and much more. NCBI contains the information regarding the Per2 gene. Currently there are 538 genes that can be found under the Per2 gene and 1,781 proteins. To answer this scientific question I used the protein sequence of the Per2 gene in humans, mice and cattle found on the NCBI database website.https://www.ncbi.nlm.nih.gov/


### Scientific Hypothesis: If there is a difference in the protein sequences of the Per2 gene between mice and humans, then the Per2 gene will be expressed differently in tissues. 

>To answer the scientific question and test my hypothesis I gathered data through NCBI database, as mentioned above. I downloaded the protein sequence of the Per2 gene for homo sapiens, house mice and cattle. The protein sequence was downloaded as a 'FASTA' file. A FASTA file is a file that is in a text-based format for representing nucleotide sequences or protein sequences in which the base pairs or amino acids are represented by individual letters https://zhanglab.ccmb.med.umich.edu/FASTA/. In order to compare the two protein sequences, I performed a pairwise sequence alignment between the human and mice. In order to determine that the mice and human sequence were different, I also did a pairwise sequence analysis with the protein sequence of cattle, which would represent the control variable in the bioinformatics test. In order to visualize the differnces between the sequences and the expression of the Per2 gene in different tissues between the mouse and human, a scatterplot was created to depict different tissues and their transcript expression.

### Part 2: Load the data and perform the bioinformatics analysis (Pairwise Sequence Alignment)

>A pairwise sequence aligment is useful to identify whether different nuclei acid or protein sequences have areas of similarity. In the following pairwise sequence alignment, the house mouse and human protein sequence for the Per2 gene were aligned and sequenced. The code read a FASTA file which was downloaded from NCBI. After running the code, the pairise sequence alignment results in a score which identifies how similar the two sequences are.

>After performing this bioinformatics method on both variables from the hypothesis, both the mouse and human protein sequence were compared to a control variable which in this case was the cattle. For my control varible I chose a mammal that is phenotypically different from both humans and mice. 

```python
# Reading in the protein sequences of mice and humans

seq1=next(SeqIO.parse("/Users/marianas/Desktop/Project_3/proteinHP.faa","fasta"))
seq2=next(SeqIO.parse("/Users/marianas/Desktop/Project_3/proteinMP.faa","fasta"))

# Pairwise sequence alignment between both sequences
alignments=pairwise2.align.globalxx(seq1.seq, seq2.seq)

# Lengths printed for seq 1 and 2 and the pairwise sequence alignment
print(len(seq1))
print(len(seq2))
print(len(alignments))

#Print the pairwise sequence alignment and score
for a in alignments:
    print(format_alignment(*a))
```

```python
# Read in the sequence of human and cattle

seq1=next(SeqIO.parse("/Users/marianas/Desktop/Project_3/proteinHP.faa","fasta"))
seq3=next(SeqIO.parse("/Users/marianas/Desktop/Project_3/proteinCP.faa","fasta"))

# Pairwise sequence alignment betwen cattle and human
control_align=pairwise2.align.globalxx(seq1.seq, seq3.seq)

#lenghts printed for each sequence and alignment
print(len(seq1))
print(len(seq3))
print(len(control_align))

# Print the alignment and score
for a in control_align:
    print(format_alignment(*a))
```

```python
# Read sequence between mouse and cattle

seq2=next(SeqIO.parse("/Users/marianas/Desktop/Project_3/proteinMP.faa","fasta"))
seq3=next(SeqIO.parse("/Users/marianas/Desktop/Project_3/proteinCP.faa","fasta"))

# Pairwise sequence alignment between mouse and cattle
control2_align=pairwise2.align.globalxx(seq2.seq, seq3.seq)

# Print lengths of each sequence and alignment
print(len(seq2))
print(len(seq3))
print(len(control2_align))

# Print alignment and score
for a in control2_align:
    print(format_alignment(*a))
```

The code shows the division between the human vs mouse pairwise sequence alignment score dividied by the longest sequence. ### Create a scatterplot showing the expression of the Per2 gene in different tissues

>Here we are using matplotlib to visualize the expression of the Per2 gene in both the mouse and human tissues. The x-axis is labeled as tissues and lists the different tissues in which the Per2 gene is expressed in mice and humans. The y-axis is labeled as RPKM which is a unit of transcript expression. Scatterplots are a good way to visualize whether or not two or more variables have a correlation or relationship. In this case the blue dots represent the transcript expression of humans in different tissues and the orange dots represent the transcript expression of mice. The data was gathered from the NCBI database which has information on the gene expression https://www.ncbi.nlm.nih.gov/gene/18627The and https://www.ncbi.nlm.nih.gov/gene/8864 . 

```python
# plotting Per2 expression in different tissues


# Data: x represents the tissues and y1 and y2 represent the RPKM of mice and humans
x=["Adrenal","Bladder","Brain","Colon","Duodenum","Fat","Heart","Kidney","Liver","Lung","Ovary","Placenta","Small Intestine","Spleen","Stomach","Testis"]
y1=[1.92,3.075,4.041,4.424,2.263,5.826,2.254,2.45,3.697,4.9,4.567,1.591,3.577,2.714,4.252,8.586]
y2=[15.261,12.767,2.82,5.962,2.206,4.664,2.11,2.548,1.89,6.482,7.092,3.022,2.288,2.089,1.405,1.683]




#Multiple line plots 
plt.scatter(x,y1,label="RPKM_Human")
plt.scatter(x,y2,label="RPKM_Mouse")
plt.xlabel('Tissue')
plt.ylabel('RPKM')
plt.title('Per2 Gene Tissue Expression in Humans and Mice')
plt.tight_layout()
plt.xticks(rotation=90)


# show legend
plt.legend()

# show graph
plt.show()
```

### Analysis of the results 

>The plot above shows the expression of the Per2 gene in different tissues found in both mice and humans. From the scatterplot we can note that the expression is very similar in the Duodenum, Heart, and Kidney. However, all other tissues demonstrate different RPKM between the mouse and human Per2 gene. With the pairwise sequence alignment scores, we can assume that the Per2 sequences of mice and humans are no the same. The closer that the score is to 1 then the more similar. The score for mice and human Per2 sequences was around 0.8 while the scores between mouse vs cattle and cattle vs human were around 0.71 and 0.74, respectively. Although the mouse and human pairwise sequence score was closest to 1, the score still implies that the sequences are different and therefore will express differently in tissues of the organisms. The cattle as the control variable demonstrates that the mouse and human Per2 sequences are not the same since the scores were different between both the cattle vs mouse and the human vs cattle. Therefore, my hypothesis was correct that since the Per2 sequences of mice and human are different, the expression of the Per2 gene in tissues would also be different.
